```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt


# File to Load (Remember to Change These)
data_to_load = Path("WHO_Data/WHOMortalityDatabase_Map_a. Malignant skin melanoma_19th December 2023 23 17.csv")

#with open('WHO_Data/WHOMortalityDatabase_Map_a. Malignant skin melanoma_19th December 2023 23 17.csv', 'r') as file:
    #for _ in range(10):  # Display the first 10 lines
        #print(file.readline())

# Read the CSV file, skip the first 6 rows, specify quotechar and separator

data = pd.read_csv(data_to_load)

data_df = pd.DataFrame(data)


# Display the DataFrame
data_df.head()
```

```python
aus_data = data_df[data_df['Country Name'] == 'Australia']

aus_data = aus_data[aus_data['Age group code'] != 'Age_all']

aus_data
```

```python
aus_data = aus_data.sort_values(by='Number', ascending=False)


data_m = aus_data[aus_data['Sex'] == 'Male']

data_f = aus_data[aus_data['Sex'] == 'Female']

aus_data
```

```python
data_m
```

```python
data_f
```

```python
df = pd.DataFrame(data_m)

plt.figure(figsize=(10, 6))

grouped_data = df.groupby('Year')['Number'].sum()

plt.plot(grouped_data.index, grouped_data.values, marker='o', linestyle='-', color='b')

plt.title('Number of Deaths Over Years Males')
plt.xlabel('Year')
plt.ylabel('Number of Deaths')
plt.grid(True)
plt.show()
```

```python
df = pd.DataFrame(data_f)

plt.figure(figsize=(10, 6))

grouped_data = df.groupby('Year')['Number'].sum()

plt.plot(grouped_data.index, grouped_data.values, marker='o', linestyle='-', color='b')

plt.title('Number of Deaths Over Years Females')
plt.xlabel('Year')
plt.ylabel('Number of Deaths')
plt.grid(True)
plt.show()
```
