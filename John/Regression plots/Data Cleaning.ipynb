```python
#import WHO dataset
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from scipy.stats import linregress


WHO_df = pd.read_csv('WHO_Data/WHOMortalityDatabase_Map_a. Malignant skin melanoma_19th December 2023 23 17.csv', skiprows = 5, index_col=False)
# headers = ["Region Code", "Region Name", "Country Code", "Country Name", "Year", "Sex", "Age group code", "Age Group", "Number", "Percentage of cause-specific deaths out of total deaths", 
#            "Age-standardized death rate per 100 000 standard population", "Death rate per 100 000 population"]
WHO_df.head()
```

```python
WHO_df = WHO_df[WHO_df['Age group code'] != 'Age_all']

WHO_df = WHO_df[WHO_df['Sex'] != 'All']



Aus = WHO_df[WHO_df['Country Name'] == 'Australia']

Aus
```

```python
df = pd.DataFrame(Aus)

plt.figure(figsize=(10, 6))
grouped_data = df.groupby('Year')['Number'].sum()

coe = np.polyfit(grouped_data.index, grouped_data.values, 1)
poly = np.poly1d(coe)
plt.plot(grouped_data.index, poly(grouped_data.index), color='r', label='Regression Line')

equation = f'y = {coe[0]:.2f}x + {coe[1]:.2f}\nR = {r_value:.2f}'
plt.annotate(equation, xy=(grouped_data.index.min(), poly(grouped_data.index.min())), xytext=(0.05, 0.95),
             textcoords='axes fraction', fontsize=12, color='r', ha='left', va='top')

plt.plot(grouped_data.index, grouped_data.values, marker='o', linestyle='-', color='b')

# Setting plot title and labels
plt.title(f'Number of Deaths Over Years Attributed to Melanoma in Australia\nR = {r_value:.2f}')
plt.xlabel('Year')
plt.ylabel('Number of Deaths')
plt.grid(True)
plt.show()
```
